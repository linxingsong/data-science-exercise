Problem: how to reduce car accident on some section of road from government, driver, and insurance company.

Car accident is happen everywhere. It is not only make people lose their money, it also take people's life. 

In every traffic accident, people are injured in many ways. They will lose money, health and even lives. And third parties will also participate in traffic accidents, such as insurance companies.

If we can find out the road sections where traffic accidents happen frequently and the reason why it happen so frequently, the police department can provide more preventive method to lower the accident based on the causes of these accidents. For example, government can install speed test device on the road that frequent car accident due to speeding. Government also able to put more specified warning sign on the road to get driver attention.  Driver will get notice and drive safe.  

Insurance company can have better way to calculate the car insurance fee to their customer. If people often drive on the high car accident road, insurance company will add this factor to their formula to get precise insurance cost. When accident happen, insurance company lose less money. 

The driver pay high insurance fee, they want to reduce it on next billing cycle.  They will drive more carefully.  For the long term to lower the car accident rate.
