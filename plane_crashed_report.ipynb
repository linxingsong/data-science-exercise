```python
# creat column that store all location as country
# column should stay after existed column, location.
```

```python
# get countrys 
def find_country(row):
    if ',' in row['location']:
        loc, cou = row['location'].rsplit(',', 1)
    else:
        cou = row['location']
    return cou
```

```python
if 'country' not in planeCrash.columns:
    planeCrash.insert(loc=4, column='country', value='0')
    
planeCrash['country'] = planeCrash.apply(find_country, axis = 1 )
planeCrash.head()
```

```python
planeCrash.replace('?', np.nan, inplace=True)
planeCrash.isnull().sum() / planeCrash.shape[0] * 100
```

```python
# two column fatalities and aboard contains 0 Nan value. 
# able to creat new column that calculate survivor rate by (aboard-fatalities)/ aboard.

planeCrash.loc[101]
```

```python
def cal_surRate(row):
    board, detail1 = row['aboard'].split(" ", 1)
    died, detail2 = row['fatalities'].split(" ", 1)
    
    if board is np.nan or board=='?' or board =='0':
        return np.nan
    return (int(board)-int(died))/int(board)
    
```

```python
if 'survivalRate' not in planeCrash.columns:
    planeCrash.insert(loc=13, column='survivalRate', value='0')
planeCrash['survivalRate'] = planeCrash.apply(cal_surRate, axis =1)
planeCrash.head()
```
