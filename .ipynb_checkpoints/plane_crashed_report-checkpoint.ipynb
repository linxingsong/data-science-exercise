```python
# I would like to make all date in one formates. 
# insert a column at begining, call conv_date
# convert everys date in the column, date, into conv_date.
```

```python
if 'conv_date' not in planeCrash.columns:
    planeCrash.insert(loc=0, column='conv_date', value='0')

planeCrash['conv_date'] = planeCrash.date.apply(
    lambda x: pd.to_datetime(x).strftime('%m/%d/%Y')
)
planeCrash.head()
```

```python
# creat column that store all location as country
# column should stay after existed column, location.
```

```python
# get countrys 
def find_country(row):
    if ',' in row['location']:
        loc, cou = row['location'].rsplit(',', 1)
    else:
        cou = row['location']
    return cou
```

```python
if 'country' not in planeCrash.columns:
    planeCrash.insert(loc=4, column='country', value='0')
    
planeCrash['country'] = planeCrash.apply(find_country, axis = 1 )
planeCrash.head()
```

```python
planeCrash.replace('?', np.nan, inplace=True)
planeCrash.isnull().sum() / planeCrash.shape[0] * 100
```

```python
# two column fatalities and aboard contains 0 Nan value. 
# able to creat new column that calculate survivor rate by (aboard-fatalities)/ aboard.

planeCrash.loc[101]
```

```python
def cal_surRate(row):
    board, detail1 = row['aboard'].split(" ", 1)
    died, detail2 = row['fatalities'].split(" ", 1)
    
    if board is np.nan or board=='?' or board =='0':
        return np.nan
    return (int(board)-int(died))/int(board)
    
```

```python
if 'survivalRate' not in planeCrash.columns:
    planeCrash.insert(loc=13, column='survivalRate', value='0')
planeCrash['survivalRate'] = planeCrash.apply(cal_surRate, axis =1)
planeCrash.head()
```

```python
# +++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++

# Draw graph base on year
# how many plane crash in each year
```

```python
def get_year(row):
    return pd.to_datetime(row['conv_date']).year
```

```python
if 'year' not in planeCrash.columns:
    planeCrash.insert(loc=0, column='year', value='0')
    
planeCrash['year'] = planeCrash.apply(get_year, axis = 1 )
```

```python
year_count = planeCrash['year'].groupby(planeCrash['year']).count()
```

```python
plt.plot(year_count)
plt.title('number of plane crashed by year')
plt.xlabel('Year')
plt.ylabel('count')
plt.show()
```

```python
# change to historgram
year_count.plot.bar(figsize=(200,80))
plt.title('number of plane crashed by year',fontsize = 500)
plt.xlabel('Year',fontsize = 100 )
plt.ylabel('count',fontsize = 100)
plt.xticks(fontsize = 100)
plt.yticks(fontsize = 100)
plt.show()
```

```python
operation_count = planeCrash['operator'].groupby(planeCrash['operator']).count()
```

```python
operation_count.keys()
```

```python
plt.pie(operation_count, labels = operation_count.keys(), autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

```python
# unable to see. 
# try group all opertaor who have less than 3 plant crash as one group, 3 to 10 as second group, 
# and rest are independes groups
```

```python
dic = operation_count.to_dict()
cleaned_dic = {'lower10': 0, 'btw11to20': 0}
count=0
count1 =0
for key in dic:
    if dic[key] <= 10:
        count += dic[key]
    elif dic[key] > 10 and dic[key]<=20:
        count1 += dic[key]
    else:
        cleaned_dic[key]=dic[key]
cleaned_dic['lower10']=count
cleaned_dic['btw11to20']=count1
```

```python
# convert back to dataframe
cleaned_dic
```

```python
oper_df = pd.DataFrame([cleaned_dic], columns=cleaned_dic.keys()).transpose()
oper_df
```

```python
fig = plt.figure(figsize=[8,8])
 
ax = fig.add_subplot(111)
ax.pie(oper_df, labels = oper_df.index ,autopct='%1.1f%%',
        shadow=False, startangle=90,pctdistance=1.1, labeldistance=1.2)
ax.axis('equal')

plt.show()
```
